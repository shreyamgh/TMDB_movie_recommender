```python
#extracting the relevent columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
#dropping the null values as number of null values were insignificant to effect the results
movies.dropna(inplace=True)
```

```python
#we will have to merge the geners, keywords, cast and crew columns in overview column 
#for that we will first have to make genres, keywords, cast and crew columns in more redable format
movies.iloc[0].genres
```

```python
#as the genres column is in form of string of list so we will have to make it in list format
#for that we will create a helper function, which will extract the name from the list(ex- Action) and append it into a new list
#in ast module there is a function literal_eval convert string of list into list
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
movies.dropna(inplace=True)
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['genres']
```

```python
#repeating the same process for keywords column
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#for cast column we have to extract top 3 cast i.e first 3 dict from the cast dict
#Assuming first three are the top three
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
#movies['cast'] = movies['cast'].apply(convert3)
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

```python
#we need to fetch just the director name only from the crew, we create a function for that 
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
#overview column is in string form we need to convert it into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#we need to remove the spaces from the cast and crew columns so that our recommender system doesnt get counfused over same first or last name
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
#concatinating all below columns into new tag column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df
```

```python
#tags columns is in list, we have to convert it into string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#convertig the tags into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

## As the tags is in text format so to find the similarity between tags we need to perform vectorization. Each movie will turn out to be a vector. So when a user will choose a movie to recommend, other movies we will return the closest vectors to the choosen movie.Textvectorization have several techniqes the one we are going to perform is called bag of words.In bag of words we will combine all the tags, then will extract the top lets say 50000 most frequent words.Then from the most frequent words we will pick each word and see how many times that word have appeared in a particular movie and create a table of that and repeat the process for each movie. So each row of the table is our vector.While performing the vectorization we will not consider the stop words(i.e. are and or to from etc).In python we have a library to perform the vectorization.

```python
#importing libraries to perform steming
import nltk
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
#creating object of the PorterStemmer class
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names()
```

The output we were getting were words like dance, dancing, dancer but we need dance dance dance for that we have to perform steming. Which i will perform above and that will return the desired output 

## Now to find the similarity, we need to calculate the distance of each movie with every other movie, the less the distance the greater the similarity. And we won't calculate the eucludian distance instead we will calculate the cosine distance(i.e the angle between the vectors). We have function in scikit learn for this called cosine_similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Batman')
```
